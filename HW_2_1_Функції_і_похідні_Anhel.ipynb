1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x') #оголошення символічної змінної
f_x = 10*x - 5 #оголошення функції
derivative_f_x = sp.diff(f_x,x) #обчилення похідної
print(f"Похідна функції 10x-5:", derivative_f_x) #виведення значення
```

**Варіант 1.** Як я розумію, можна вручну записати функцію похідної, аби не переводити символьне значення в нумерік, бо з цим виникає проблема


```python
x_values = np.linspace(0, 10, 100)
f = 10*x - 5
df = np.full_like(x_values, 10) #у нас 10 фактично константа, тому потрібно використати np.full_like()
plt.figure(figsize=(6, 3))
plt.plot(x_values, f, label='f(x) = 10x-5', color='blue')
plt.plot(x_values, df, label="f'(x) = 10", color='green')

plt.title("Оцінка та витрачені години")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

**Варіант 2.** Можна також з допомогою lambdify.

```python
print(f_x, derivative_f_x)
```

```python
f_x_func = sp.lambdify(x, f_x, 'numpy')
derivative_f_x_values = np.full_like(x_values, float(derivative_f_x)) #тут значення похідної фактично є константою, тому треба записувати ось так
```

```python
x_values = np.linspace(0, 10, 100)
f_x_values = f_x_func(x_values) #значення f(x) для кожного х
```

```python
plt.figure(figsize=(6, 3))
plt.plot(x_values, f_x_values, label='f(x) = 10x-5', color='blue')
plt.plot(x_values, derivative_f_x_values, label="f'(x) = 10", color='green')

plt.title("Оцінка та витрачені години")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

**Висновки:** похідна більша 0, тобто зростання годин навчання позитивно впливає на оцінку. При цьому похідна f'(x)=10. Тобто кожна додаткова година навчання приносить 10 балів. Це фактично видно і на графіку, що кожна додаткова година приносить 10 балів.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
f_sin = sp.sin(x)
derivative_f_sin = sp.diff(f_sin)
print("Похідна sin(x):", derivative_f_sin)
```

```python
f_sin_func = sp.lambdify(x, f_sin, 'numpy')
derivative_f_sin_func = sp.lambdify(x, derivative_f_sin, 'numpy')
```

```python
x_values = np.linspace(0, 10, 100)
f_sin_values = f_sin_func(x_values)
derivative_f_sin_func = derivative_f_sin_func(x_values)
```

```python
plt.figure(figsize=(4, 2))
plt.plot(x_values, f_sin_values, label='f(x)=sin(x)', color='blue')
plt.plot(x_values, derivative_f_sin_func, label="f'(x) = cos(x)", color='green')

plt.title("f(x)=sin(x)")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
x = sp.symbols('x')
f_ln = sp.ln(x)
derivative_f_ln = sp.diff(f_ln)
print("Похідна ln(x):", derivative_f_ln)
```

```python
f_ln_func = sp.lambdify(x, f_ln, 'numpy')
derivative_f_ln_func = sp.lambdify(x, derivative_f_ln, 'numpy')
```

```python
x_values = np.linspace(1, 10, 100) #поставив від 1, бо не можна ділити на 0, викидає помилку
f_ln_values = f_ln_func(x_values)
derivative_f_ln_func = derivative_f_ln_func(x_values)
```

```python
plt.figure(figsize=(4, 2))
plt.plot(x_values, f_ln_values, label='f(x)=ln(x)', color='blue')
plt.plot(x_values, derivative_f_ln_func, label="f'(x) = 1/x", color='green')

plt.title("f(x)=ln(x)")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
